```python
import jax.numpy as np # imports numpy for vector-matrix math
import numpy.random as np_random
import numpy as onp
from jax.nn import softmax
from jax import grad, jit, vmap, jacfwd, partial # all the derivative and gradients
from tqdm import tqdm, trange
# %matplotlib notebook
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
@jit
def f(x, u):
    p, xdot = np.split(x, 2)
    pnew = p + xdot*0.1
    xdotnew = 0.9*xdot + u * 0.1
    return np.concatenate([pnew, xdotnew])
# get the jacobians 
dfdx = jit(jacfwd(f,0))
dfdu = jit(jacfwd(f,1))

@jit
def p(x):
    x1 = x-np.array([0.7,0.])
    x2 = x+np.array([0.5,0.5])
    x3 = x-np.array([-0.4,0.6])
    return np.exp(-0.5*np.dot(x1,x1)/0.01) \
                + np.exp(-0.5*np.dot(x2,x2)/0.01) \
                + np.exp(-0.5*np.dot(x3,x3)/0.01)

@jit
def q(s, x): # x is traj, s is sample
    if len(x.shape) > 1:
        _in = -0.5 * np.sum(np.square(s-x[:,:2]), axis=1)/0.1
    else:
        _in = -0.5 * np.sum(np.square(s-x[:2]))/0.1
    return np.mean(np.exp(_in))

def barr_fun(x):
    _val = 0.
    if np.abs(x[0]) > 1:
        _val += 10*(np.abs(x[0])-1)**2
    if np.abs(x[1]) > 1:
        _val += 10*(np.abs(x[1])-1)**2
    return _val

gradbarr = grad(barr_fun)
gradq = grad(q, 1)

def forward_pass(x0, u):
    x = []
    x_t = x0.copy()
    for t, u_t in enumerate(u):
        x.append(x_t)
        x_t = f(x_t, u_t)
    return x

def backward_pass(x, u, hist, step_size=1e-2, N=20):
    u_star = []
    rho = x[-1]*0.
    x_tot = np.stack(x+hist)
    s = np_random.uniform(-1, 1., size=(N, 2))
    log_q_vals = np.log(vmap(q, in_axes=(0, None))(s, x_tot)+1e-2)
    log_p_vals = np.log(vmap(p)(s)+1e-2)
    log_q_vals -= np.max(log_q_vals)
    log_q_vals += 1e-3
    log_p_vals -= np.max(log_p_vals)
    w = log_p_vals-log_q_vals
    w -= np.max(w)
    w = np.exp(w)
    for x_t, u_t in zip(x[::-1], u[::-1]):
        rho = -np.dot(w, vmap(gradq, in_axes=(0, None))(s, x_t)) + np.dot(dfdx(x_t, u_t).T, rho) + gradbarr(x_t)
#         du =  0.01*u_t + np.dot(dfdu(x_t, u_t).T, rho)
#         u_star.append(u_t + step_size * du)
        u_star.append(-np.dot(dfdu(x_t, u_t).T, rho))
        print(rho)
    return u_star[::-1]


```

```python
# some parameters 
max_iter  = 100
tf        = 5
dt        = 0.1 # time step
N         =  int(tf/dt)
threshold = 0.001
step_size = 1e-2

# initialize the control with a random sequence of actions
u = [np.array([-0.1, -0.1]) for t in range(N)]
x0 = np.array([1.,2.]) # initial condition
```

```python
trajectories = []
costs = []
with tqdm(total=max_iter) as pbar:
    for k in range(max_iter):
        x, cost, x_f = forward_pass(x0, u )
        u = backward_pass(x, u, x_f,  step_size)
        x, cost, x_f = forward_pass(x0, u)
        pbar.set_description('Current Cost: {}'.format(cost))
        pbar.update()
        if k % 20 == 0:
            trajectories.append(x)
            costs.append(cost)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))
for i, x in enumerate(trajectories):
    x = np.stack(x)
    axs[0].plot(x[:,0], x[:,1], label='iter {}'.format(i*10))
#     axs[0].plot(x[:,0], label='iter {}'.format(i*10))
#     axs[0].plot(x[:,1], label='iter {}'.format(i*10))
# axs[0].set_xlim(-2,2)
# axs[0].set_ylim(-2,2)
axs[1].plot(np.stack(costs), label='iter {}'.format(i*10))
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('x-y trajectory')
axs[0].legend()
axs[1].set_xlabel('iteration x 10 ')
axs[1].set_ylabel('cost')
axs[1].set_title('cost over iteration')
plt.show()
```
